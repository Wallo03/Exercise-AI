```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('https://frenzy86.s3.eu-west-2.amazonaws.com/python/data/Startup.csv')
df
```

```python
df.corr() #crea un dataframe con tutti i valori a 1
```

```python
sns.heatmap(df.corr(), annot=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
y = df['Profit']
#y = df.drop(['R&D Spend', 'Administration', 'Marketing Spend'], axis=1)
y
```

```python
X = df[['R&D Spend', 'Administration', 'Marketing Spend']]
X
```

```python
## 2 suddividere il problema in Training e Test

X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size = 0.2, 
                                                    random_state = 0
                                                    )
```

```python
X_train.shape, X_test.shape #quante righe e quante colonne hanno
```

```python
## 3 Creare ed allenare il modell (fit) sulla parte di training
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
## 4 Creare la predizione sulla parte di TEST

res_df = pd.DataFrame(data=list(zip(y_pred, y_test)),columns=['predicted', 'real'])
res_df
```

```python
res_df['error'] = res_df['real'] - res_df['predicted']
res_df
```

```python
length = y_pred.shape[0] #  
x = np.linspace(0,length,length)

plt.figure(figsize=(10,7))
plt.plot(x, y_test, label='real y')
plt.plot(x, y_pred, label="predicted y'")
plt.legend(loc=2)
```

```python
## 5 misurare l'errore del mio modello con magari l'r2 score
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

r2score = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)

print('R2_score: ', r2score)
print('MAE: ', mae)
print('MSE: ', mse)
print('RMSE: ', rmse)
```
